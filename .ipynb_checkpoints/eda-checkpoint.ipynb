# EDA Notebook

## Imports & Setup

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
from datasets import load_dataset
import warnings
warnings.filterwarnings('ignore')

sns.set_theme(style='whitegrid', palette='muted', font_scale=1.2)
plt.rcParams['figure.dpi'] = 120

LABEL_NAMES = ['BACKGROUND', 'OBJECTIVE', 'METHODS', 'RESULTS', 'CONCLUSIONS']
COLORS = sns.color_palette('muted', 5)

print('Ready.')
```

## Load the Dataset

```python
ds = load_dataset('armanc/pubmed-rct20k')

train_df = ds['train'].to_pandas()
val_df   = ds['validation'].to_pandas()
test_df  = ds['test'].to_pandas()

print(ds)
print(f'\nTrain : {len(train_df):,}')
print(f'Val   : {len(val_df):,}')
print(f'Test  : {len(test_df):,}')
```

## Quick Look at Data

```python
print(train_df.head(10).to_string())
print('\nDtypes:\n', train_df.dtypes)
print('\nNull check:\n', train_df.isnull().sum())
print('\nUnique labels:', train_df['label'].unique())
```

## Class Distribution

```python
fig, axes = plt.subplots(1, 3, figsize=(16, 5))

for ax, (df, split) in zip(axes, [(train_df, 'Train'), (val_df, 'Val'), (test_df, 'Test')]):
    counts = df['label'].value_counts().reindex([l.lower() for l in LABEL_NAMES])
    ax.bar(counts.index, counts.values, color=COLORS)
    ax.set_title(f'{split} ({len(df):,} sentences)')
    ax.set_xlabel('Label')
    ax.set_ylabel('Count')
    ax.tick_params(axis='x', rotation=30)
    for i, v in enumerate(counts.values):
        ax.text(i, v + 200, f'{v:,}', ha='center', fontsize=9)

plt.suptitle('Class Distribution Across Splits', fontsize=14, y=1.02)
plt.tight_layout()
plt.savefig('figures/class_distribution.png', bbox_inches='tight')
plt.show()

# Print exact percentages
print('=== Train class breakdown ===')
for label, count in train_df['label'].value_counts().items():
    print(f'  {label:<15} {count:>7,}  ({count/len(train_df)*100:.1f}%)')
```

## Sentence Length Analysis

```python
for df in [train_df, val_df, test_df]:
    df['token_count'] = df['text'].apply(lambda x: len(x.split()))

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Distribution by split
for df, split in [(train_df, 'Train'), (val_df, 'Val'), (test_df, 'Test')]:
    axes[0].hist(df['token_count'], bins=50, alpha=0.5, label=split)
axes[0].set_title('Sentence Length Distribution')
axes[0].set_xlabel('Word Count')
axes[0].set_ylabel('Frequency')
axes[0].legend()

# Length by label (train only)
train_df.boxplot(column='token_count', by='label', ax=axes[1],
                 boxprops=dict(color='steelblue'))
axes[1].set_title('Sentence Length by Label (Train)')
axes[1].set_xlabel('Label')
axes[1].set_ylabel('Word Count')
axes[1].tick_params(axis='x', rotation=30)
plt.suptitle('')

plt.tight_layout()
plt.savefig('figures/sentence_lengths.png', bbox_inches='tight')
plt.show()

print('=== Token count stats (train) ===')
print(train_df.groupby('label')['token_count'].describe().round(1))
print(f'\nOverall 95th percentile: {train_df["token_count"].quantile(0.95):.0f} tokens')
print(f'Overall max            : {train_df["token_count"].max()} tokens')
```

## Sample Sentences Per Class

```python
print('=== Sample sentences per class ===\n')
for label in ['background', 'objective', 'methods', 'results', 'conclusions']:
    samples = train_df[train_df['label'] == label]['text'].sample(3, random_state=42).tolist()
    print(f'--- {label.upper()} ---')
    for i, s in enumerate(samples, 1):
        print(f'  {i}. {s}')
    print()
```

## EDA Summary & Key Findings

### Dataset Overview
- 235,892 total sentences across train (176,642), validation (29,672), and test (29,578) splits
- 4 features: `abstract_id`, `label`, `text`, `sentence_id`
- No missing values in any split

### Class Imbalance
- Significant imbalance present: `methods` (33.6%) and `results` (32.8%) dominate
- `objective` is the minority class at only 7.8% of training data
- Distribution is consistent across all three splits — no data leakage concerns
- **Implication:** Overall accuracy alone will be misleading. Per-class F1 is the primary evaluation metric.

### Sentence Length
- Sentences are short — median word count is 22–26 tokens across all classes
- 95th percentile is 55 tokens, max is 296 (outlier results sentences)
- **Implication:** `max_len = 128` is safe for both the LSTM and DistilBERT — captures 95%+ of sentences with no wasted compute

### Text Characteristics
- Numbers are anonymized as `@` throughout — consistent across all classes
- Each class has distinguishable linguistic patterns:
  - `background`: general medical context, prior work references
  - `objective`: purpose-driven language ("The purpose of this study was to...")
  - `methods`: procedural language, heavy use of `@` (anonymized numeric values)
  - `results`: statistical language, dense `@` usage, longest/most variable sentences
  - `conclusions`: recommendation language ("should", "further studies are necessary")

### Modeling Implications
- Per-class F1 should be tracked, especially for `objective` (minority class)
- `max_len = 128` for both LSTM and DistilBERT tokenization
- The `@` token should be kept — it carries signal, especially in `methods` and `results`
- Class weights may be needed during training to handle imbalance
